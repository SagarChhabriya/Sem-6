### Simple Linear Regression Task 2: Predicting packages on the basis of cgpa using LinearRegression 

#### Step 1: Import Libraries and Load data

```python
# import libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error 
from sklearn.linear_model import LinearRegression 
```

```python
# Load Data 
df = pd.read_csv(r"A:\Code\PY\DSc\datasets\01-simple-linear-reg\cgpa.csv")
df
```

#### Step 2: Perform EDA + Preprocessing + Train Test Split


```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(16,4))

df["cgpa"].plot(kind="kde",ax=ax1)
df['package'].plot(kind="kde",ax=ax2)

ax3.set_title("Relation between CGPA & Package")
sns.scatterplot(data=df, x="cgpa", y="package",ax=ax3)
```

```python
# Train test split

X_train, X_test, Y_train, Y_test = train_test_split(df["cgpa"],df["package"],random_state=42, test_size=0.2)
```

```python
X_train = X_train.values.reshape(-1,1) # convert pandas series to numpy array and then reshape the numpy array to 2d from 1d
Y_train = Y_train.values.reshape(-1,1)
X_test = X_test.values.reshape(-1,1)
Y_test = Y_test.values.reshape(-1,1)
```

```python
import numpy as np 
np.random.randint(1,10,size=6).reshape(2,3)
```

#### Step 3: Train the model


```python
# Create object
lr = LinearRegression()

# fit the data
lr.fit(X_train, Y_train)
```

#### Step 4: Predict and Evaluate


```python
y_pred = lr.predict(X_test)

print("R2 Score: ", r2_score(Y_test, y_pred))
print("Mean Square Error: ", mean_squared_error(Y_test, y_pred))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train, lr.predict(X_train),color="red")
plt.show()
```

#### Step 5: Find Y_intercept and Slope


```python

m = lr.coef_
c = lr.intercept_
x = 6.89 # Try with new value
y = m * x + c 
y
```
