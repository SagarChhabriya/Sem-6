1. **Merge two DataFrames on a single column and print the resulting DataFrame.**
   - **DataFrame 1 (`df1`):**
     ```
     ID    | Name     | Age
     ----------------------
     1     | Alice    | 25
     2     | Bob      | 30
     3     | Charlie  | 35
     ```
   - **DataFrame 2 (`df2`):**
     ```
     ID    | City      | Country
     --------------------------
     1     | New York  | USA
     2     | London    | UK
     4     | Tokyo     | Japan
     ```
   - **Task:** Merge `df1` and `df2` on the `ID` column and display the result.



````python
import pandas as pd
````

````python
df1 = pd.DataFrame({
    "ID":[1,2,3],
    "Name":["Alice","Bob","Charlie"],
    "Age":[25,30,35]
})

df2 = pd.DataFrame({
    "ID":[1,2,4],
    "City":["New York","London","Tokyo"],
    "Country":["USA","UK","Japan"]
})
````

````python
pd.merge(df1,df2,on="ID",how="outer") 
````

2. **Perform an inner join between two DataFrames on a common key.**
   - **DataFrame 1 (`df1`):**
     ```
     Product_ID | Product_Name | Price
     -------------------------------------
     101        | Laptop       | 1000
     102        | Phone        | 500
     103        | Tablet       | 300
     ```
   - **DataFrame 2 (`df2`):**
     ```
     Product_ID | Stock
     -------------------
     101        | 50
     102        | 100
     104        | 20
     ```
   - **Task:** Perform an inner join on `Product_ID` and display the merged DataFrame.



````python
df1 = pd.DataFrame({
    "Product_ID":[101,102,103],
    "Product_Name":["Laptop","Phone","Tablet"],
    "Price":[1000,500,300]
})

df2 = pd.DataFrame({
    "Product_ID":[101,102,104],
    "Stock":[50,100,20]
})

pd.merge(df1,df2,on="Product_ID",how="inner")
````

3. **Merge two DataFrames using a left join and show the result.**
   - **DataFrame 1 (`df1`):**
     ```
     Customer_ID | Customer_Name | Email
     ----------------------------------------
     1           | John Doe      | john@example.com
     2           | Jane Smith    | jane@example.com
     3           | Sarah Lee     | sarah@example.com
     ```
   - **DataFrame 2 (`df2`):**
     ```
     Customer_ID | Purchase_Amount
     ----------------------------
     1           | 200
     2           | 150
     4           | 300
     ```
   - **Task:** Merge `df1` and `df2` using a left join on `Customer_ID` and print the result.



````python
df1 = pd.DataFrame({
    "Customer_ID":[1,2,3],
    "Customer_Name":["John Doe","Jane Smith","Sarah Lee"],
    "Email":["john@example.com","jane@example.com","sarah@example.com"]
})
df2 = pd.DataFrame({
    "Customer_ID":[1,2,4],
    "Purchase_Amount":[20,150,300]
})

pd.merge(df1,df2,on="Customer_ID",how="left")
````
